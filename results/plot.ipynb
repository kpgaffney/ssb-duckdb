# SSB on DuckDB

```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt

sns.set_theme(style="whitegrid")

df = pd.read_csv('results.csv')

for scale_factor in [1, 3, 10, 30, 100]:
    df_i = df[(df['Scale Factor'] == scale_factor) & (df['Trial'] != 0)]
    if not df_i.empty:
        plt.figure(figsize=(10, 4))
        sns.barplot(data=df_i, x='Query', y='Time', hue='Variant')
        plt.ylabel('Time (s)')
        plt.title(f'Scale Factor = {scale_factor}')
        plt.savefig(f'sf-{scale_factor}.pdf')
```
